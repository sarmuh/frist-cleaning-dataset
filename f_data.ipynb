```python
# Datasetni yuklab olish
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df = pd.read_csv(url)

# Ma'lumotlarni ko'rish
df.head(5)
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
df['Has-Cabin'] = np.where(df['Cabin'].isna(), 0, 1)
df.head(10)
```

```python
df.drop('Cabin', axis=1,inplace=True)
```

```python
df.isnull().sum()
```

```python
df[df['Embarked'].isnull()]
```

```python
df.dropna(subset='Embarked',inplace=True)
```

```python
df[df['Embarked'].isnull()]
```

```python
df.isnull().sum()
```

```python
imputer = SimpleImputer(strategy='mean')

age_data = df[['Age']]

df['Age'] = imputer.fit_transform(age_data)
```

```python
df.isnull().sum()
```

```python
le = LabelEncoder()

df['Sex'] = le.fit_transform(df['Sex'])

df.head(10)
```
